### Q1. What is Flask Framework? What are the advantages of Flask Framework?

Flask is a pythonic library to create API's and expose the functions/ methods to the outerworld so that it can be accessed from anywhere.A code which is created on Pyhton to be accessed through website, URL, pool, etc.

Advantages of Flask Framework:

Simplicity: Flask is known for its simplicity and ease of use. It has a small codebase and minimalistic design, making it easy to understand and get started with. Flask follows the "micro" approach, meaning it provides only the essential tools for web development without any unnecessary overhead.

Flexibility: Flask provides a flexible and modular design that allows developers to choose and integrate various third-party libraries and tools based on their specific requirements. This gives developers the freedom to design their application architecture and choose components as per their preferences.


Pythonic: Flask is built on Python, which is a popular and powerful programming language known for its simplicity and readability. This makes Flask a preferred choice for Python developers, as it allows them to leverage their existing Python skills and knowledge while building web applications.


Testing: Flask provides built-in support for testing, making it easy to write and execute unit tests, integration tests, and end-to-end tests for web applications. This helps in ensuring the quality and reliability of the application during development.

Large Community and Documentation: Flask has a large and active community of developers, which means there is ample support available in the form of documentation, tutorials, forums, and online resources. This makes it easier for developers to learn, troubleshoot, and get help when needed.

Overall, Flask is a popular web framework due to its simplicity, flexibility, extensibility, Pythonic nature, lightweightness, testing support, and a large community, making it a preferred choice for many web developers.

App routing in Flask refers to the process of mapping URLs (Uniform Resource Locators) to specific functions or view handlers in a Flask application. When a user makes a request to a specific URL, Flask uses routing to determine which function or view handler should handle that request and generate an appropriate response.

In Flask, app routing is achieved using decorators that are applied to view functions. These decorators define the URL pattern that the view function should respond to.
